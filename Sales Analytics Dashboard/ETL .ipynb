```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\PC-Mariem\Downloads\Sample - Superstore.csv\Sample - Superstore.csv",
    encoding='latin1'
)
df.head()
```

```python
# Aperçu des premières lignes
df.head()

# Taille du dataset
print("Shape :", df.shape)

# Liste des colonnes
print("\nColonnes :")
print(df.columns)

# Types des colonnes
print("\nTypes des variables :")
print(df.dtypes)

# Statistiques descriptives
print("\nStats numériques :")
print(df.describe())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Vérification
df[['Order Date', 'Ship Date']].head()
```

```python
print("Valeurs manquantes :")
df.isnull().sum()
```

```python
# Nombre de doublons
print("Doublons :", df.duplicated().sum())

# Suppression doublons
df = df.drop_duplicates()
```

```python
df['Sales'] = df['Sales'].astype(float)
df['Quantity'] = df['Quantity'].astype(int)
df['Discount'] = df['Discount'].astype(float)
df['Profit'] = df['Profit'].astype(float)
```

```python
df.to_csv(r"C:\Users\PC-Mariem\Downloads\superstore_clean.csv", index=False)
print("Fichier exporté avec succès !")
```
